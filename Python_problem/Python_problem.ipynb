```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret) 
auth.set_access_token(access_key, access_secret) 
api = tweepy.API(auth) 
tweets = api.user_timeline(screen_name='midasIIITD')
```

```python
with open('data.jsonl','w') as r:
  for i in tweets:
    json.dump(i._json,r)
    r.write('\n')
```

```python
rows = []
for tweet in tweets:
  row = [] 
  row.append(tweet.text)
  date = tweet.created_at.strftime('%m/%d/%Y')
  row.append(date)
  time = tweet.created_at.strftime('%H:%M')
  row.append(time)
  row.append(tweet.favorite_count)
  row.append(tweet.retweet_count)
  img_list = ''
  if 'media' in tweet.entities.keys():
        for med in tweet.entities['media']:
            if med['type'] == "photo":
                img_list = img_list +" "+ med["media_url"]
        row.append(img_list)
  else:
        row.append('-')
  rows.append(row)

data = pd.DataFrame(rows,columns = ['tweet_text','Date','Time','Likes','Retweets','images'])# converting the list of data rows into a DataFrame
```
